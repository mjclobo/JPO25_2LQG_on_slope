# Defining model parameters

```julia

dev = GeophysicalFlows.CPU()

Ny = Nx = n = 1024                # 2D resolution = n²


###################################################################################
## Geometry and background fields
###################################################################################

Nz = nlayers = 2                 # number of layers
f0 = f₀= 8.3e-5            # Coriolis param [s^-1]
g = 9.81                    # gravity
H = [2000., 2000.]     # the rest depths of each layer;

Us = [0.01, 0.017634058790440166, 0.024]
rhos = 1025.75 - 0.5775

kt = 0.

rho0 = 1025.        # Boussinesq reference density

# always zero for now, model can't even take this in as a parameter...
V = zeros(nlayers)


# running model

rho = ρ = [0., 1025.75]
rho[1] = ρ[1] = rhos[1]

strat_str = "uni_strat"
shear_str = "uni_shear"

# change U here
U = [0.,0.]
U[1] = Us[1]

gprime = (g/rho0)*(ρ[2] - ρ[1])

Ld = sqrt(gprime * sum(H)) / 2 / f0

L = Lx = Ly = 25*2*pi*Ld   # domain size [m]

###################################################################################
## Planetary PV gradient
###################################################################################
betas = βs = collect(range(-1.5,1.5,13)) * (U[1]/2) * Ld^-2
β = 0. # 1e-11              # the y-gradient of planetary PV; beta = β = 0. * (U[1]/2) * Ld^-2

###################################################################################
## Topography
###################################################################################

# set topography w/ current options:
topo_type = "y_slope"

# Defining a bulk slope parameter...should probably change this!

# setting slope(s)
S32 = f0 * U[1] / gprime

dyetab_over_S32 = collect(range(-1.5,1.5,13))
h0s = dyetab_over_S32 .* S32
# h0s = h0s[7:end]

# tuple of (bottom slope, h_rms)
h0 = (h0s[1], 0.)    #

# central wavenumber for rough topo
kt = 0.

# defining GFJL topo params
interim_params = mod_params(topo_type=topo_type, h0=h0, kt=kt)

topographic_pv_gradient, eta = define_topo(interim_params)

###################################################################################
# Dissipation
###################################################################################

# biharmonic viscosity
nν = 4
νstar = 0. # 10^-13
ν = νstar * ((U[1]-U[end])/2) * (Lx/2/pi)^7

dyn_nu = false

# Linear bottom drag
μstars = [0.125, 0.25, 0.5, 1.0, 2.0, 4.0]
mus = @. μstars * ((U[1]-U[end]) / 2) / Ld

# Quadratic bottom drag
κstars = [0.]
kappas = [0.] # @. κstars / Ld

###################################################################################
## Time stepping and sample rate
###################################################################################

if ν==0.
    stepper = "FilteredETDRK4"
    # af = 0.
    # global filt_order=4
    # global innerK = 2/3
else
    stepper = "ETDRK4"
    # af = 1/3
    # global filt_order=4
    # innerK = 2/3
end

# should I set dt via CFL condition instead?
dt = 600.

###################################################################################
## Details of saving data
###################################################################################
global diags = diag_bools(two_layer_kspace_modal_nrg_budget_bool=true);

# where to save model output; frequency of output
data_dir = "/scratch/cimes/ml1994/QG/julia/data/sloping_LD_two_layer_data/"

alt_data_dir = "/scratch/cimes/ml1994/QG/julia/data/JFM_May2025_data/"

f_s = 5. * (Ld / ((U[1]-U[end])/2)) #  N eddy turnover periods

# sampling period in time steps (set above)
nsubs = round(Int64, f_s/dt)

###################################################################################
## Model run parameters
###################################################################################

# steady-state  parameters
ss_yr_max = ceil(100 * (Ld / ((U[1]-U[end])/2)) / 3600 / 24 / 365.25)  # number of eddy periods converted to years for model run
yr_increment = 1.0 # how often to write to a save file

restart_bool = false

pre_buoy_restart_file = true
data_dir_pre_buoy = "/scratch/cimes/ml1994/QG/julia/data/two_layer_drag_study/"
data_dir_2L = "/scratch/cimes/ml1994/QG/julia/data/sloping_2L_hi_output_WFS/"

###################################################################################
## Define params struct
###################################################################################
global model_params = mod_params(
data_dir = data_dir,
Nz = Nz, Nx = Nx, Ny = Ny, Lx = Lx, Ly = Ly, Ld = Ld,
H = H,
rho0 = rho0, rho = rho, strat_str = strat_str,
shear_str = shear_str, U = U,
μ = mus[1], κ = kappas[1], nν = nν, ν = ν, dyn_nu = dyn_nu,
eta = eta, topographic_pv_gradient = topographic_pv_gradient, topo_type = topo_type, h0 = h0,
f0=f0, β = β,
dt = dt,
stepper = stepper,
dev = dev,
restart_bool = restart_bool,
pre_buoy_restart_file = pre_buoy_restart_file,
data_dir_pre_buoy = data_dir_pre_buoy,
ss_yr_max = ss_yr_max,
yr_increment = yr_increment,
nsubs = nsubs)

###################################################################################
## Initialize model
###################################################################################

prob, prob_filt = initialize_model(model_params);

# prob = set_initial_conditions(prob, prob_filt, model_params)

sol, clock, params, vars, grid = prob.sol, prob.clock, prob.params, prob.vars, prob.grid;


```

```julia
using Roots

fig, ax = plt.subplots(3,1, figsize=(5,10))

fig.tight_layout(pad=0.0)

ax1=ax[1]; ax2=ax[2]; ax3=ax[3];

lsize = 20
tsize = 12
fsize = 20
bsize = 16



x = h0s ./ S32
x2 = x[7:end-1]

############################################################################

############################################################################
vs_ind = 1

# vel scales are twelve
# total BT, U BT, V BT, total BC, U BC, V BC, (7) total L1, U L1, V L1, total L2, U L2, V L2

############################################################################

############################################################################
# function dispersion_relation(L, beta_topo, U1, mode_string)

#     if mode_string=="BT"
#         sf = -sign(beta_topo) * 1
#     elseif mode_string=="BC"
#         sf = -sign(beta_topo) * -1
#     end
    
#     k2 = 1 * (2 * pi / L)^2   # pre-factor of 2 assumes isotropic length scale;
    
#     A = k2 * (k2 + Ld^-2)
#     B = beta_topo * (k2 + (2*Ld^2)^-1) - U1 * k2 * (k2 + Ld^-2)
#     C = U1 * k2 * (U1 * (2 * Ld^2)^-1 - beta_topo)

#     dr = -B / 2 / A + sf * real(sqrt(Complex(B^2 - 4 * A * C))) / 2 / A

#     # dr = topographic wave from LaCasce (1998) ?

#     return dr
# end

function dispersion_relation_full_bottom(L, beta_topo, U1, mode_string, mu_loc)
    
    k2 = 1 * (2 * pi / L)^2   # pre-factor of 2 assumes isotropic length scale;
    
    # A = k2 * (k2 + Ld^-2)
    # B = beta_topo * (k2 + (2*Ld^2)^-1) - U1 * k2 * (k2 + Ld^-2)
    # C = U1 * k2 * (U1 * (2 * Ld^2)^-1 - beta_topo)

    A = k2 * (k2 + Ld^-2)
    B = (beta_topo - im * mu_loc * sqrt(k2)) * (k2 + (2*Ld^2)^-1) - U1 * k2 * (k2 + Ld^-2)
    C = U1 * k2 * (U1 * (2 * Ld^2)^-1 - beta_topo + im * mu_loc * sqrt(k2))
    
    
    # if mode_string=="BT"
    #     dr = U1/2 - (beta_topo / 2) / k2
    # elseif mode_string=="BC"
    #     dr = U1/2 - (beta_topo / 2) / (k2 + Ld^-2)
    # end

    # this is full dispersion relation for bottom slope PV gradient in lower layer only
    dr_both = [- B / 2 / A + real(sqrt(Complex(B^2 - 4 * A * C))) / 2 / A; - B / 2 / A - real(sqrt(Complex(B^2 - 4 * A * C))) / 2 / A]
    if mode_string=="BT"
        indmax = argmax(abs.(dr_both))
        dr = real(dr_both[indmax])
    elseif mode_string=="BC"
        indmin = argmin(abs.(dr_both))
        dr = real(dr_both[indmin])
    end

    return dr
end

function dispersion_relation_Rhines(L, beta_topo, U1, mode_string)

    k2 = 1 * (2 * pi / L)^2   # pre-factor of 2 assumes isotropic length scale;

    if mode_string=="BT"
         c = - beta_topo / (2 * k2)
    elseif mode_string=="BC"
         c = -(beta_topo/2 + U1 * Ld^-2) / (k2 + Ld^-2)
    end
    
    # generalized Rhines scale from Theiss (2006)
    
    
    return c  # note that c is technically phase speed, not angular frequency!
end

function dispersion_relation_RG(L, beta_topo, U1; kx=0, ky=0)

    if ky==0
        k2 = 1.0 * (2 * pi / L)^2
    else
        k2 = kx^2 + ky^2
    end

    # see Wang et al. (2016) eddy phase speeds paper
    c = U1 - 0.5 * beta_topo / (k2 + 0.5 * Ld^-2)

    return c
end

function dispersion_relation_BT(L, beta_topo, U1, mode_string)

    # if mode_string=="BT"
    #     sf = 1
    # elseif mode_string=="BC"
    #     sf = -1
    # end

    # if mode_string=="BT"
    #     sf = -sign(beta_topo) * 1
    # elseif mode_string=="BC"
    #     sf = -sign(beta_topo) * -1
    # end
    
    k2 = 1.0 * (2 * pi / L)^2   # pre-factor of 1.0 assumes isotropic length scale;
    
    # this is barotropic wave form

    if mode_string=="BT"
        c = U1/2 - (beta_topo/2) * (2*k2 + Ld^-2) / (2*k2 * (k2 + Ld^-2)) - sign(beta_topo) * real(sqrt(Complex(U1^2 * k2^2 * (k2^2 - Ld^-4) + Ld^-4 * (beta_topo/2)^2)) / (2 * k2 * ( k2 + Ld^-2)))
    elseif mode_string=="BC"
        c = U1/2 - (beta_topo/2) * (2*k2 + Ld^-2) / (2*k2 * (k2 + Ld^-2)) + sign(beta_topo) * real(sqrt(Complex(U1^2 * k2^2 * (k2^2 - Ld^-4) + Ld^-4 * (beta_topo/2)^2)) / (2 * k2 * ( k2 + Ld^-2)))
    end
    # if mu_loc==mus[4]
    #     if U1^2 * k2^2 * (k2^2 - Ld^-4) + Ld^-4 * (beta_topo/2)^2 < 0
    #         println("BCI")
    #     end
    # end

    return c  # note that c is technically phase speed, not angular frequency!
end

dispersion_relation(L, beta_topo, U1, mode_string) = dispersion_relation_BT(L, beta_topo, U1, mode_string)

############################################################################

############################################################################

function define_modal_scale_ratios(mu_loc)

    LBT = zeros(length(h0s)) .* NaN
    LBT_full = zeros(length(h0s)) .* NaN
    LBT_sym = zeros(length(h0s)) .* NaN
    L_kappa_BT = zeros(length(h0s)) .* NaN
    L_kappa_BT_sym = zeros(length(h0s)) .* NaN
    L_kappa_BT_heur = zeros(length(h0s)) .* NaN
    L_ratio_BT = zeros(length(h0s)) .* NaN
    L_ratio_BT_sym = zeros(length(h0s)) .* NaN
    NLP_BT = zeros(length(h0s)) .* NaN
    NLP_BT_sym = zeros(length(h0s)) .* NaN

    if mu_loc > mus[1]
        h_start=1
    else
        h_start=7
    end
    
    for h_ind in range(h_start, length(h0s))
    
        h0 = (h0s[h_ind], 0.0)
        kappa_loc = kappas[1]
        
        global model_params = redef_mu_kappa_topoPV_h0(model_params, mu_loc, kappa_loc, (0., 0.), h0)
        global topo_PV, eta = define_topo(model_params)
        global model_params = redef_mu_kappa_topoPV_h0(model_params, mu_loc, kappa_loc, topo_PV, h0)

        if h_ind != 14
            a = load(data_dir*jld_name_2L(model_params,62.95))

            c1 = 5
            c2 = 0.36

            v_scale_BT_sym =  (sqrt(2*a["jld_data"]["two_layer_xspace_modal_nrg_budget"][1]))

            v_scale_BT_heur =  (sqrt(2*a["jld_data"]["two_layer_xspace_modal_nrg_budget"][1]) - sign(h0s[h_ind]) * 0.01/2)
            v_scale_BT =  sqrt(2*a["jld_data"]["two_layer_xspace_modal_nrg_budget"][1])
            v_scale_BC = sqrt(2*a["jld_data"]["two_layer_xspace_modal_nrg_budget"][2])

            ############################################################################
            # vel scales are twelve
            # total BT, U BT, V BT, total BC, U BC, V BC, total L1, U L1, V L1, total L2, U L2, V L2
            
            # vBT_ind = 3
            # vBC_ind = 6
            # if mu_loc==mus[2]
            #     u1_scale = vel_scales_kap0p25[h_ind,7]
            #     u2_scale = vel_scales_kap0p25[h_ind,10]
            #     v_scale_BT = vel_scales_kap0p25[h_ind,vBT_ind]
            #     v_scale_BC = vel_scales_kap0p25[h_ind,vBC_ind]
            # elseif mu_loc==mus[3]
            #     u1_scale = vel_scales_kap0p5[h_ind,7]
            #     u2_scale = vel_scales_kap0p5[h_ind,10]
            #     v_scale_BT = vel_scales_kap0p5[h_ind,vBT_ind]
            #     v_scale_BC = vel_scales_kap0p5[h_ind,vBC_ind]
            # elseif mu_loc==mus[4]
            #     u1_scale = vel_scales_kap1[h_ind,7]
            #     u2_scale = vel_scales_kap1[h_ind,10]
            #     v_scale_BT = vel_scales_kap1[h_ind,vBT_ind]
            #     v_scale_BC = vel_scales_kap1[h_ind,vBC_ind]
            # elseif mu_loc==mus[5]
            #     u1_scale = vel_scales_kap2[h_ind,7]
            #     u2_scale = vel_scales_kap2[h_ind,10]
            #     v_scale_BT = vel_scales_kap2[h_ind,vBT_ind]
            #     v_scale_BC = vel_scales_kap2[h_ind,vBC_ind]
            # elseif mu_loc==mus[6]
            #     u1_scale = vel_scales_kap4[h_ind,7]
            #     u2_scale = vel_scales_kap4[h_ind,10]
            #     v_scale_BT = vel_scales_kap4[h_ind,vBT_ind]
            #     v_scale_BC = vel_scales_kap4[h_ind,vBC_ind]
            # end

            # v_scale_BT_sym = v_scale_BT

            alpha = (f0 * h0s[h_ind] / H[2])

            L_kappa_BT[h_ind] = 2 * pi * abs(v_scale_BT) / mu_loc
            L_kappa_BT_heur[h_ind] = 2 * pi * abs(v_scale_BT_heur) / mu_loc
            L_kappa_BT_sym[h_ind] = 2 * pi * abs(v_scale_BT_sym) / mu_loc

            # println(h_ind)

            # f_BT = L0 -> sqrt(2*a["jld_data"]["two_layer_xspace_modal_nrg_budget"][1]) - abs(dispersion_relation(L0, alpha, 0.01, "BT"))
            # f_BT_full = L0 -> sqrt(2*a["jld_data"]["two_layer_xspace_modal_nrg_budget"][1]) - abs(dispersion_relation_full_bottom(L0, alpha, 0.01, "BT", mu_loc))
            # f_BT_sym = L0 -> sqrt(2*a["jld_data"]["two_layer_xspace_modal_nrg_budget"][1]) - abs(dispersion_relation(L0, alpha, 0.0, "BT")) 

            f_BT = L0 -> v_scale_BT - abs(dispersion_relation(L0, alpha, 0.01, "BT"))
            f_RGBT = L0 -> v_scale_BT - abs(dispersion_relation_RG(L0, alpha, 0.01))
            f_BT_full = L0 -> v_scale_BT - abs(dispersion_relation_full_bottom(L0, alpha, 0.01, "BT", mu_loc))
            f_BT_sym = L0 -> v_scale_BT - abs(dispersion_relation(L0, alpha, 0.0, "BT")) 

            NLP_BT[h_ind] = v_scale_BT / (abs(alpha) * Ld^2)
            NLP_BT_sym[h_ind] = v_scale_BT_sym / (abs(alpha) * Ld^2)
            
            if h_ind != 7
                # LBT[h_ind] = abs(find_zero(f_BT, L_kappa_BT[h_ind]))
                # if f_BC(L_kappa_BC[h_ind])!=Inf

                try
                    # BT full
                    LBT[h_ind] = abs(find_zero(f_BT, L_kappa_BT[h_ind]*5)) # abs(find_zero(f_BC, [0., 1e8])) #

                    LBT_full[h_ind] = abs(find_zero(f_BT_full, L_kappa_BT[h_ind]*5))
                    
                    LBT_sym[h_ind] = abs(find_zero(f_BT_sym,L_kappa_BT_sym[h_ind])) 

                    # RG BT
                    # LBT[h_ind] = abs(find_zero(f_RGBT, L_kappa_BT[h_ind]*0.05))

                    if mu_loc==mus[6]
                        println(LBT[h_ind])
                    end

                    # LBT_sym[h_ind] = 2 * pi * sqrt(abs((sqrt(2*a["jld_data"]["two_layer_xspace_modal_nrg_budget"][1]) - sign(alpha)*0.01/2)/(alpha/2)))
                catch
                    println("whoops")
                else
                    # LBC[h_ind] = 1/0
                end

            else
                LBT[h_ind] = 1/0
                LBT_full[h_ind] = 1/0
                LBT_sym[h_ind] = 1/0
            end

            
            if LBT[h_ind] / L_kappa_BT[h_ind] < 1
                L_ratio_BT[h_ind] = grid.kr[argmax(abs.(a["jld_data"]["two_layer_kspace_modal_nrg_budget"][:,7]))] * LBT[h_ind]
            else
                L_ratio_BT[h_ind] = grid.kr[argmax(abs.(a["jld_data"]["two_layer_kspace_modal_nrg_budget"][:,7]))] * L_kappa_BT[h_ind]
            end

            if LBT_sym[h_ind] / L_kappa_BT_sym[h_ind] < 1
                L_ratio_BT_sym[h_ind] = grid.kr[argmax(abs.(a["jld_data"]["two_layer_kspace_modal_nrg_budget"][:,8]))] * LBT_sym[h_ind]
            else
                L_ratio_BT_sym[h_ind] = grid.kr[argmax(abs.(a["jld_data"]["two_layer_kspace_modal_nrg_budget"][:,8]))] * L_kappa_BT_sym[h_ind]
            end
            
        end
        
    end
        
    return LBT, LBT_full, LBT_sym, L_kappa_BT, L_kappa_BT_sym, L_kappa_BT_heur, L_ratio_BT ./ (2*pi), L_ratio_BT_sym ./ (2*pi), NLP_BT, NLP_BT_sym
end

# LBT_mu0, LBT_full_mu0, LBT_sym_mu0, LKBT_mu0, LKBT_sym_mu0, LKBT_heur_mu0, Lrat_BT_mu0, Lrat_BT_sym_mu0, NLP_BT_mu0, NLP_BT_sym_mu0 = define_modal_scale_ratios(mus[1])

LBT_mu1, LBT_full_mu1, LBT_sym_mu1, LKBT_mu1, LKBT_sym_mu1, LKBT_heur_mu1, Lrat_BT_mu1, Lrat_BT_sym_mu1, NLP_BT_mu1, NLP_BT_sym_mu1 = define_modal_scale_ratios(mus[2])

LBT_mu2, LBT_full_mu2, LBT_sym_mu2, LKBT_mu2, LKBT_sym_mu2, LKBT_heur_mu2, Lrat_BT_mu2, Lrat_BT_sym_mu2, NLP_BT_mu2, NLP_BT_sym_mu2 = define_modal_scale_ratios(mus[3])

LBT_mu3, LBT_full_mu3, LBT_sym_mu3, LKBT_mu3, LKBT_sym_mu3, LKBT_heur_mu3, Lrat_BT_mu3, Lrat_BT_sym_mu3, NLP_BT_mu3, NLP_BT_sym_mu3 = define_modal_scale_ratios(mus[4])

LBT_mu4, LBT_full_mu4, LBT_sym_mu4, LKBT_mu4, LKBT_sym_mu4, LKBT_heur_mu4, Lrat_BT_mu4, Lrat_BT_sym_mu4, NLP_BT_mu4, NLP_BT_sym_mu4 = define_modal_scale_ratios(mus[5])

LBT_mu5, LBT_full_mu5, LBT_sym_mu5, LKBT_mu5, LKBT_sym_mu5, LKBT_heur_mu5, Lrat_BT_mu5, Lrat_BT_sym_mu5, NLP_BT_mu5, NLP_BT_sym_mu5 = define_modal_scale_ratios(mus[6])



############################################################################

############################################################################

ax1.axhline(1, color="red", linestyle="dashed")

# ax1.plot(x, LBT_mu1, color="k", marker="^", linestyle="dotted", label=L"\kappa^* = 0.25")
# ax1.plot(x, LBT_mu2, color="k", marker="d", linestyle="-", label=L"\kappa^* = 0.5")
# ax1.plot(x, LBT_mu3, color="k", marker="s", linestyle="--", label=L"\kappa^* = 1.0")
# ax1.plot(x, LBT_mu4, color="k", marker="o", linestyle="-.", label=L"\kappa^* = 2.0")
# ax1.plot(x, LBT_mu5, color="k", marker="x", linestyle="dotted", label=L"\kappa^* = 4.0")

eh = 1

ax1.plot(x[6:end-1], (LKBT_sym_mu1[6:end-1] ./ LBT_sym_mu1[6:end-1]).^eh, color="k", marker="^", linestyle="dotted", label=L"\kappa^* = 0.25")
ax1.plot(x[5:end], (LKBT_sym_mu2[5:end] ./ LBT_sym_mu2[5:end]).^eh, color="k", marker="d", linestyle="-", label=L"\kappa^* = 0.5")
ax1.plot(x, (LKBT_sym_mu3 ./ LBT_sym_mu3).^eh, color="k", marker="s", linestyle="--", label=L"\kappa^* = 1.0")
ax1.plot(x, (LKBT_sym_mu4 ./ LBT_sym_mu4).^eh, color="k", marker="o", linestyle="-.", label=L"\kappa^* = 2.0")
ax1.plot(x, (LKBT_sym_mu5 ./ LBT_sym_mu5).^eh, color="k", marker="x", linestyle="dotted", label=L"\kappa^* = 4.0")


ax1.set_ylabel(L" L^{\kappa}_\mathrm{BT}  \, / \, L_\mathrm{BT}^\mathrm{Rh} ", fontsize=lsize)

ax1.set_xlim(x[1], x[end])

ax1.set_ylim(-0.05, 3)
ax1.grid()

ax1.legend(fontsize=tsize, ncols=2, bbox_to_anchor=(0.825, 1.335), loc="upper right")
ax1.tick_params(labelsize=tsize, which="both")
ax1.set_xticklabels([])

ax2.axhline(1, color="red", linestyle="dashed")

# # ax2.plot(x2, LS_nu0_kap0p125[1:end-1,2] ./ LS_nu0_kap0p125[1:end-1,1], color="k", marker=".", linestyle="-.", label=L"\kappa^* = 0.125")
# ax2.plot(x, LBT_sym_over_LBT_mu1, color="k", marker="^", linestyle="dotted", label=L"\kappa^* = 0.25")
# ax2.plot(x, LBT_sym_over_LBT_mu2, color="k", marker="d", linestyle="-", label=L"\kappa^* = 0.5")
# ax2.plot(x, LBT_sym_over_LBT_mu3, color="k", marker="s", linestyle="--", label=L"\kappa^* = 1.0")
# ax2.plot(x, LBT_sym_over_LBT_mu4, color="k", marker="o", linestyle="-.", label=L"\kappa^* = 2.0")
# ax2.plot(x,LBT_sym_over_LBT_mu5, color="k", marker="x", linestyle="dotted", label=L"\kappa^* = 4.0")


ax2.plot(x[6:end-1], (LKBT_mu1[6:end-1] ./ LBT_mu1[6:end-1]).^eh, color="k", marker="^", linestyle="dotted", label=L"\kappa^* = 0.25")
ax2.plot(x[5:end], (LKBT_mu2[5:end] ./ LBT_mu2[5:end]).^eh, color="k", marker="d", linestyle="-", label=L"\kappa^* = 0.5")
ax2.plot(x, (LKBT_mu3 ./ LBT_mu3).^eh, color="k", marker="s", linestyle="--", label=L"\kappa^* = 1.0")
ax2.plot(x, (LKBT_mu4 ./ LBT_mu4).^eh, color="k", marker="o", linestyle="-.", label=L"\kappa^* = 2.0")
ax2.plot(x, (LKBT_mu5 ./ LBT_mu5).^eh, color="k", marker="x", linestyle="dotted", label=L"\kappa^* = 4.0")


# ax2.plot(x[6:end-1], (LKBC_mu1[6:end-1] ./ LBC_mu1[6:end-1]).^eh, color="gray", marker="^", linestyle="dotted", label=L"\kappa^* = 0.25")
# ax2.plot(x[5:end], (LKBC_mu2[5:end] ./ LBC_mu2[5:end]).^eh, color="gray", marker="d", linestyle="-", label=L"\kappa^* = 0.5")
# ax2.plot(x, (LKBC_mu3 ./ LBC_mu3).^eh, color="gray", marker="s", linestyle="--", label=L"\kappa^* = 1.0")
# ax2.plot(x, (LKBC_mu4 ./ LBC_mu4).^eh, color="gray", marker="o", linestyle="-.", label=L"\kappa^* = 2.0")
# ax2.plot(x, (LKBC_mu5 ./ LBC_mu5).^eh, color="gray", marker="x", linestyle="dotted", label=L"\kappa^* = 4.0")


ax2.set_ylabel(L" L^{\kappa}_\mathrm{BT} \, / \, L_\mathrm{BT}^\mathrm{Sh}  ", fontsize=lsize)
# ax3.set_xlabel(L"\partial_y \eta_b / S_{3/2}", fontsize=lsize)
ax2.set_xlabel(L"\partial_y \eta_b / S_{3/2}", fontsize=lsize)

ax2.grid()

ax2.tick_params(labelsize=tsize, which="both")

# ax1.set_yscale("log")

ax2.set_xlim(x[1], x[end])


ax2.set_ylim(-0.05, 3)

# Lkappa over Ltopo

############################################################################

############################################################################


ax3.plot(x[6:end-1], (LKBT_mu1[6:end-1] ./ LBT_full_mu1[6:end-1]).^eh, color="k", marker="^", linestyle="dotted", label=L"\kappa^* = 0.25")
ax3.plot(x[5:end], (LKBT_mu2[5:end] ./ LBT_full_mu2[5:end]).^eh, color="k", marker="d", linestyle="-", label=L"\kappa^* = 0.5")
ax3.plot(x, (LKBT_mu3 ./ LBT_full_mu3).^eh, color="k", marker="s", linestyle="--", label=L"\kappa^* = 1.0")
ax3.plot(x, (LKBT_mu4 ./ LBT_full_mu4).^eh, color="k", marker="o", linestyle="-.", label=L"\kappa^* = 2.0")
ax3.plot(x, (LKBT_mu5 ./ LBT_full_mu5).^eh, color="k", marker="x", linestyle="dotted", label=L"\kappa^* = 4.0")


ax3.set_ylabel(L" L^{\kappa}_\mathrm{BT} \, / \, L_\mathrm{BT}^\mathrm{Sh}  ", fontsize=lsize)
# ax3.set_xlabel(L"\partial_y \eta_b / S_{3/2}", fontsize=lsize)
ax3.set_xlabel(L"\partial_y \eta_b / S_{3/2}", fontsize=lsize)

ax3.grid()

ax3.tick_params(labelsize=tsize, which="both")

# ax1.set_yscale("log")

ax3.set_xlim(x[1], x[end])


ax3.set_ylim(-0.05, 3)


############################################################################

############################################################################

# xmax_ind = 45
# x = grid.kr .* Ld


# for h_ind in vcat(collect(range(1,6)), 7, collect(range(8,13))) # length(h0s))
#     if h_ind<7
#         lt = "dashed"
#     elseif h_ind>=7
#         lt="solid"
#     end

#     h0 = (h0s[h_ind], 0.0)
#     kappa_loc = kappas[1]
#     mu_loc = mus[5]
    
#     global model_params = redef_mu_kappa_topoPV_h0(model_params, mu_loc, kappa_loc, (0., 0.), h0)
#     global topo_PV, eta = define_topo(model_params)

#     a = load(data_dir*jld_name_2L(model_params,62.95))

#     ax3.plot(x, a["jld_data"]["two_layer_kspace_modal_nrg_budget"][:,7] .* (a["jld_data"]["two_layer_vBT_scale"]^-3 / Ld), label=string(round(h0s[h_ind] / S32, sigdigits=3)), color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)

# end



# ax3.set_xlabel(L" | \boldsymbol{k} | \cdot \lambda", fontsize=fsize)

# ymins = [ -0.5,]
# ymaxes = ymins .+ 0.5

# ax3.grid()
# ax3.tick_params(labelsize=tsize)
# ax3.set_xlim(x[1], x[xmax_ind])

# ax3.set_ylim(-0.4, 0.1)

# plot_info_box(ax3, L"\ \ \widehat{D}_\mathrm{BC}", 0.7, 0.9, "BC", bsize)

# # ax3.set_ylim(ax3.get_ylim()[1],1.2)

############################################################################

############################################################################


ax1.text(0.075, 0.9, L"\mathbf{(a)}", fontsize=fsize, horizontalalignment="center",verticalalignment="center", transform = ax1.transAxes);
ax2.text(0.075, 0.9, L"\mathbf{(b)}", fontsize=fsize, horizontalalignment="center",verticalalignment="center", transform = ax2.transAxes);


############################################################################

############################################################################


# savefig("./drag_figs/length_scale_pred_BT.pdf",bbox_inches="tight")
```

```julia
using Roots

fig, ax = plt.subplots(4,2, figsize=(5,12))
ax1=ax[1]; ax2=ax[2]; ax3=ax[3]; ax4=ax[4];
ax5=ax[5]; ax6=ax[6]; ax7=ax[7]; ax8=ax[8];
fig.tight_layout(pad=0.0)

lsize = 20
tsize = 12
fsize = 20
bsize = 16



############################################################################

############################################################################
# vel scales are twelve
# total BT, U BT, V BT, total BC, U BC, V BC, (7) total L1, U L1, V L1, total L2, U L2, V L2

############################################################################

############################################################################
# function dispersion_relation(L, beta_topo, U1, mode_string)

#     if mode_string=="BT"
#         sf = -sign(beta_topo) * 1
#     elseif mode_string=="BC"
#         sf = -sign(beta_topo) * -1
#     end
    
#     k2 = 1 * (2 * pi / L)^2   # pre-factor of 2 assumes isotropic length scale;
    
#     A = k2 * (k2 + Ld^-2)
#     B = beta_topo * (k2 + (2*Ld^2)^-1) - U1 * k2 * (k2 + Ld^-2)
#     C = U1 * k2 * (U1 * (2 * Ld^2)^-1 - beta_topo)

#     dr = -B / 2 / A + sf * real(sqrt(Complex(B^2 - 4 * A * C))) / 2 / A

#     # dr = topographic wave from LaCasce (1998) ?

#     return dr
# end

function dispersion_relation_full_bottom(L, beta_topo, U1, mode_string, mu_loc; kx=0, ky=0, frequency=false)

    if ky==0

        k2 = 1 * (2 * pi / L)^2   # pre-factor of 2 assumes isotropic length scale; prefactor of 1 assumes zonal wavenumber only, l=0
        
        A = k2 * (k2 + Ld^-2)
        B = (beta_topo + im * mu_loc * sqrt(k2)) * (k2 + (2*Ld^2)^-1) - U1 * k2 * (k2 + Ld^-2)
        C = U1 * k2 * (U1 * (2 * Ld^2)^-1 - beta_topo - im * mu_loc * sqrt(k2))

        # a = @. k2 * (k2 + Ld^-2)
        # b = @. - U[1] * k2 * (k2 + Ld^-2) + (k2 + 0.5 * Ld^-2) * ((f0/H[2]) * h0 + 1im * μ * sqrt(k2))
        # c = @. - U[1] * k2 * ((f0/H[2]) * h0 - 0.5 * Ld^-2 *U[1] + 1im * μ * sqrt(k2)) 
        
    elseif kx==0
        
        k2 = ky^2   # pre-factor of 2 assumes isotropic length scale; prefactor of 1 assumes zonal wavenumber only, l=0
        
        A = k2 * (k2 + Ld^-2)
        B = (beta_topo) * (k2 + (2*Ld^2)^-1) - U1 * k2 * (k2 + Ld^-2)
        C = U1 * k2 * (U1 * (2 * Ld^2)^-1 - beta_topo)
    else
        k2 = kx^2 + ky^2   # pre-factor of 2 assumes isotropic length scale; prefactor of 1 assumes zonal wavenumber only, l=0
        
        A = k2 * (k2 + Ld^-2)
        B = (beta_topo + im * mu_loc * k2 / kx) * (k2 + (2*Ld^2)^-1) - U1 * k2 * (k2 + Ld^-2)
        C = U1 * k2 * (U1 * (2 * Ld^2)^-1 - beta_topo - im * mu_loc * k2/kx)
    end
    
    # if mode_string=="BT"
    #     dr = U1/2 - (beta_topo / 2) / k2
    # elseif mode_string=="BC"
    #     dr = U1/2 - (beta_topo / 2) / (k2 + Ld^-2)
    # end

    # this is full dispersion relation for bottom slope PV gradient in lower layer only
    dr_both = [- B / 2 / A + real(sqrt(Complex(B^2 - 4 * A * C))) / 2 / A; - B / 2 / A - real(sqrt(Complex(B^2 - 4 * A * C))) / 2 / A]
    if mode_string=="BT"
        indmax = 1 # argmax(abs.(dr_both))
        dr = real(dr_both[indmax])
    elseif mode_string=="BC"
        indmin = 2 #argmin(abs.(dr_both))
        dr = real(dr_both[indmin])
    end

    # if real(B^2 - 4 * A * C) < 0
    #     println(B^2 - 4 * A * C)
    # end

    dr_both_comp = sqrt(k2) .* [- B / 2 / A + sqrt(Complex(B^2 - 4 * A * C)) / 2 / A; - B / 2 / A - sqrt(Complex(B^2 - 4 * A * C)) / 2 / A]

    disc = imag.(dr_both_comp)
    if any(abs.(disc) .> 0)
        max_growth = maximum((disc))
    else
        max_growth = 0.
    end

    # if (B^2 - 4 * A * C) > 0
    #     println(sqrt(B^2 - 4 * A * C) / B)
    # end
    # dr = U1/2 - beta_topo * (2*k2 + Ld^-2) / (2*k2 * (k2 + Ld^-2)) + sf * real(sqrt(Complex(U1^2 * k2^2 * (k2^2 - 2 * Ld^-2) + 2 * Ld^-2 * beta_topo^2)) / (2 * k2 * ( k2 + Ld^-2)))


    return dr, max_growth

end

############################################################################

############################################################################

function define_modal_scale_ratios(mu_loc; mu_bool=false)

    Nell = 5000
    
    c_BT_full = zeros(length(h0s), Nell) .* NaN
    c_BC_full = zeros(length(h0s), Nell) .* NaN
    growth = zeros(length(h0s), Nell)
    neg_max = ones(length(h0s))

    # ks = collect(range(0.001, 2, Nell)) ./ Ld
    ells = collect(range(0, 100*2*pi*Ld, Nell-1))

    ells = [ells;Inf]

    for h_ind in range(1, length(h0s))
    
        h0 = (h0s[h_ind], 0.0)
        kappa_loc = kappas[1]
        
        global model_params = redef_mu_kappa_topoPV_h0(model_params, mu_loc, kappa_loc, (0., 0.), h0)
        global topo_PV, eta = define_topo(model_params)
        global model_params = redef_mu_kappa_topoPV_h0(model_params, mu_loc, kappa_loc, topo_PV, h0)

        if mu_bool==true
            mu_LSA = mu_loc
        else
            mu_LSA = 0.
        end

        if h_ind != 14
            a = load(data_dir*jld_name_2L(model_params,62.95))

            CBCh = a["jld_data"]["two_layer_kspace_modal_nrg_budget"][:,4]

            if sum(findall(x -> x<0, CBCh[1:20])) >0
                neg_max[h_ind] = maximum(findall(x -> x<0, CBCh[1:20]))
            end

            v_scale_BT =  (sqrt(2*a["jld_data"]["two_layer_xspace_modal_nrg_budget"][1]))

            alpha_s = (f0 * h0s[h_ind] / H[2])

            for (i,L0) in enumerate(ells)
                # if h_ind != 7
    
                c_BT_full[h_ind, i], growth[h_ind, i] = dispersion_relation_full_bottom(L0, alpha_s, 0.01, "BT", mu_LSA)#; kx=2*pi/L0, ky=2*pi/Lx*6)  # mu_loc)
                c_BC_full[h_ind, i], x = dispersion_relation_full_bottom(L0, alpha_s, 0.01, "BC", mu_LSA)#; kx=2*pi/L0, ky=2*pi/Lx*6)  # mu_loc)
                # c_BC_full[h_ind, i] = dispersion_relation_RG(L0, alpha_s, 0.01)
                
                # end
            end
        end
        
    end
        
    return ells, c_BT_full, c_BC_full, growth, Int.(neg_max)
end

ells, c_BT_topo, c_BC_topo, growth, neg_max = define_modal_scale_ratios(mus[2])


ells, c_BT_topo_mu, c_BC_topo_mu, growth_mu, neg_max_mu = define_modal_scale_ratios(mus[2], mu_bool=true)

############################################################################

############################################################################

max_growth_ind = zeros(Int,length(h0s))

for i in range(1, length(h0s))
    max_growth_ind[i] = Int(argmax(growth[i,:]))
end


############################################################################

############################################################################

x = (ells.^-1) .* (Ld ) * 2 * pi

k_ctopo = deepcopy(x)

# x = x.^-1

for h_ind in [1, 3, 5, 9, 11, 13, 7] # vcat(collect(range(1,6)), collect(range(8,13)), 7) # length(h0s))
    if h_ind<7
        lt = "dashed"
    elseif h_ind>=7
        lt="solid"
    end

    halting_ind = argmin(abs.(( LBT_full_mu2[h_ind]^-1 * Ld * 2 * pi) .- x))

    neg_CBC_ind = argmin(abs.(grid.kr[neg_max[h_ind]] .* Ld .- x))

    halting_ind_mu = argmin(abs.(( LBT_full_mu2[h_ind]^-1 * Ld * 2 * pi) .- x))

    neg_CBC_ind_mu = argmin(abs.(grid.kr[neg_max_mu[h_ind]] .* Ld .- x))

    if h_ind < 7

        ax1.plot(x, growth[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        
        ax2.plot(x, c_BT_topo[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        ax2.plot(x, c_BC_topo[h_ind,:], label=string(round(h0s[h_ind] / S32, sigdigits=3)), color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)


        ax3.plot(x, growth_mu[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        
        ax4.plot(x, c_BT_topo_mu[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        ax4.plot(x, c_BC_topo_mu[h_ind,:], label=string(round(h0s[h_ind] / S32, sigdigits=3)), color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)

    elseif h_ind > 7

        ax5.plot(x, growth[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        
        ax6.plot(x, c_BT_topo[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        ax6.plot(x, c_BC_topo[h_ind,:], label=string(round(h0s[h_ind] / S32, sigdigits=3)), color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)

        ax7.plot(x, growth_mu[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        
        ax8.plot(x, c_BT_topo_mu[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        ax8.plot(x, c_BC_topo_mu[h_ind,:], label=string(round(h0s[h_ind] / S32, sigdigits=3)), color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        
    else

        ax1.plot(x, growth[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        ax5.plot(x, growth[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        
        ax2.plot(x, c_BT_topo[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        ax2.plot(x, c_BC_topo[h_ind,:], label=string(round(h0s[h_ind] / S32, sigdigits=3)), color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)

        ax6.plot(x, c_BT_topo[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        ax6.plot(x, c_BC_topo[h_ind,:], label=string(round(h0s[h_ind] / S32, sigdigits=3)), color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)

        ##
        ax3.plot(x, growth_mu[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        ax7.plot(x, growth_mu[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        
        ax4.plot(x, c_BT_topo_mu[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        ax4.plot(x, c_BC_topo_mu[h_ind,:], label=string(round(h0s[h_ind] / S32, sigdigits=3)), color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)

        ax8.plot(x, c_BT_topo_mu[h_ind,:], color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
        ax8.plot(x, c_BC_topo_mu[h_ind,:], label=string(round(h0s[h_ind] / S32, sigdigits=3)), color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)
    
    end

    if h_ind > 7
        ax6.plot(x[neg_CBC_ind], c_BC_topo[h_ind,neg_CBC_ind], "+", color=h_color(h0s[h_ind],hc_range,colors))
        ax6.plot(x[halting_ind] , c_BC_topo[h_ind,halting_ind], "o", color=h_color(h0s[h_ind],hc_range,colors))

        ax8.plot(x[neg_CBC_ind_mu], c_BC_topo_mu[h_ind,neg_CBC_ind_mu], "+", color=h_color(h0s[h_ind],hc_range,colors))
        ax8.plot(x[halting_ind_mu] , c_BC_topo_mu[h_ind,halting_ind_mu], "o", color=h_color(h0s[h_ind],hc_range,colors))
        
    elseif h_ind<7
        ax2.plot(x[halting_ind] , c_BT_topo[h_ind,halting_ind], "o", color=h_color(h0s[h_ind],hc_range,colors))

        ax4.plot(x[halting_ind_mu] , c_BT_topo_mu[h_ind,halting_ind_mu], "o", color=h_color(h0s[h_ind],hc_range,colors))
        
    end

    # ax1.plot(x[max_growth_ind[h_ind]], c_BC_topo[h_ind,max_growth_ind[h_ind]], "o", color=h_color(h0s[h_ind],hc_range,colors))
end

# ax[1].set_yscale("log")
# ax[2].set_yscale("log")

ax2.set_ylim(-0.025, 0.025)
ax4.set_ylim(-0.025, 0.025)

# ax2.set_ylim(-1, 1)

ax2.set_xlim(0,2)
ax4.set_xlim(0,2)
############################################################################

############################################################################
ax2.plot(NaN,NaN,"o", color="black", label=L"L_\mathrm{topo}")
ax2.plot(NaN,NaN,"+", color="black", label=L"L_\mathrm{min} ( C_\mathrm{BC} < 0)")

ax4.plot(NaN,NaN,"o", color="black", label=L"L_\mathrm{topo}")
ax4.plot(NaN,NaN,"+", color="black", label=L"L_\mathrm{min} ( C_\mathrm{BC} < 0)")


ax2.legend()
ax2.grid()

ax4.legend()
ax4.grid()

ax6.set_ylim(-0.025, 0.025)
ax8.set_ylim(-0.025, 0.025)
# ax4.set_ylim(-1, 1)

ax6.set_xlim(0,2)
ax8.set_xlim(0,2)

ax6.grid()
ax8.grid()

ax1.set_xlim(0,2)
ax5.set_xlim(0,2)

ax3.set_xlim(0,2)
ax7.set_xlim(0,2)

ax1.grid()
ax5.grid()

ax3.grid()
ax7.grid()


# ax1.axhline(1, color="red", linestyle="dashed")

# ax1.plot(x, LBT_mu1, color="k", marker="^", linestyle="dotted", label=L"\kappa^* = 0.25")
# ax1.plot(x, LBT_mu2, color="k", marker="d", linestyle="-", label=L"\kappa^* = 0.5")
# ax1.plot(x, LBT_mu3, color="k", marker="s", linestyle="--", label=L"\kappa^* = 1.0")
# ax1.plot(x, LBT_mu4, color="k", marker="o", linestyle="-.", label=L"\kappa^* = 2.0")
# ax1.plot(x, LBT_mu5, color="k", marker="x", linestyle="dotted", label=L"\kappa^* = 4.0")

# eh = 1

# ax1.plot(x[6:end-1], c_topo_mu1, color="k", marker="^", linestyle="dotted", label=L"\kappa^* = 0.25")
# ax1.plot(x[5:end], c_topo_mu2, color="k", marker="d", linestyle="-", label=L"\kappa^* = 0.5")
# ax1.plot(x, (LKBT_sym_mu3 ./ LBT_sym_mu3).^eh, color="k", marker="s", linestyle="--", label=L"\kappa^* = 1.0")
# ax1.plot(x, (LKBT_sym_mu4 ./ LBT_sym_mu4).^eh, color="k", marker="o", linestyle="-.", label=L"\kappa^* = 2.0")
# ax1.plot(x, (LKBT_sym_mu5 ./ LBT_sym_mu5).^eh, color="k", marker="x", linestyle="dotted", label=L"\kappa^* = 4.0")


# ax1.set_ylabel(L" L^{\kappa}_\mathrm{BT}  \, / \, L_\mathrm{BT}^\mathrm{Rh} ", fontsize=lsize)

# ax1.set_xlim(x[1], x[end])

# ax1.set_ylim(-0.05, 3)
# ax1.grid()

# ax1.legend(fontsize=tsize, ncols=2, bbox_to_anchor=(0.825, 1.335), loc="upper right")
# ax1.tick_params(labelsize=tsize, which="both")
# ax1.set_xticklabels([])


############################################################################

############################################################################

# xmax_ind = 45
# x = grid.kr .* Ld


# for h_ind in vcat(collect(range(1,6)), 7, collect(range(8,13))) # length(h0s))
#     if h_ind<7
#         lt = "dashed"
#     elseif h_ind>=7
#         lt="solid"
#     end

#     h0 = (h0s[h_ind], 0.0)
#     kappa_loc = kappas[1]
#     mu_loc = mus[5]
    
#     global model_params = redef_mu_kappa_topoPV_h0(model_params, mu_loc, kappa_loc, (0., 0.), h0)
#     global topo_PV, eta = define_topo(model_params)

#     a = load(data_dir*jld_name_2L(model_params,62.95))

#     ax3.plot(x, a["jld_data"]["two_layer_kspace_modal_nrg_budget"][:,7] .* (a["jld_data"]["two_layer_vBT_scale"]^-3 / Ld), label=string(round(h0s[h_ind] / S32, sigdigits=3)), color=h_color(h0s[h_ind],hc_range,colors),linestyle=lt)

# end



# ax3.set_xlabel(L" | \boldsymbol{k} | \cdot \lambda", fontsize=fsize)

# ymins = [ -0.5,]
# ymaxes = ymins .+ 0.5

# ax3.grid()
# ax3.tick_params(labelsize=tsize)
# ax3.set_xlim(x[1], x[xmax_ind])

# ax3.set_ylim(-0.4, 0.1)

# plot_info_box(ax3, L"\ \ \widehat{D}_\mathrm{BC}", 0.7, 0.9, "BC", bsize)

# # ax3.set_ylim(ax3.get_ylim()[1],1.2)

############################################################################

############################################################################


# ax1.text(0.075, 0.9, L"\mathbf{(a)}", fontsize=fsize, horizontalalignment="center",verticalalignment="center", transform = ax1.transAxes);
# ax2.text(0.075, 0.9, L"\mathbf{(b)}", fontsize=fsize, horizontalalignment="center",verticalalignment="center", transform = ax2.transAxes);


############################################################################

############################################################################


# savefig("./drag_figs/topo_wavespeeds.pdf",bbox_inches="tight")
######################
ax[1].text(0.55, 0.95, L"\mathbf{(a)} \, : \kappa^* = 0.25, \, \beta_\mathrm{t}^* = 0.25", bbox=Dict("facecolor"=>"white", "alpha"=>0.9), fontsize=lsize, horizontalalignment="center",verticalalignment="center", transform = ax[1].transAxes);
ax[3].text(0.55, 0.95, L"\mathbf{(b)} \, : \kappa^* = 0.25, \, \beta_\mathrm{t}^* = 1.25", bbox=Dict("facecolor"=>"white", "alpha"=>0.9), fontsize=lsize, horizontalalignment="center",verticalalignment="center", transform = ax[3].transAxes);
ax[5].text(0.55, 0.95, L"\mathbf{(c)} \, : \kappa^* = 1.0, \, \beta_\mathrm{t}^* = 0.25", bbox=Dict("facecolor"=>"white", "alpha"=>0.9), fontsize=lsize, horizontalalignment="center",verticalalignment="center", transform = ax[5].transAxes);
ax[7].text(0.55, 0.95, L"\mathbf{(d)} \, : \kappa^* = 1.0, \, \beta_\mathrm{t}^* = 1.25", bbox=Dict("facecolor"=>"white", "alpha"=>0.9), fontsize=lsize, horizontalalignment="center",verticalalignment="center", transform = ax[7].transAxes);

ax[2].text(0.1, 0.95, L"\mathbf{(e)}", bbox=Dict("facecolor"=>"white", "alpha"=>0.9), fontsize=lsize, horizontalalignment="center",verticalalignment="center", transform = ax[2].transAxes);
ax[4].text(0.1, 0.95, L"\mathbf{(f)}", bbox=Dict("facecolor"=>"white", "alpha"=>0.9), fontsize=lsize, horizontalalignment="center",verticalalignment="center", transform = ax[4].transAxes);
ax[6].text(0.1, 0.95, L"\mathbf{(g)}", bbox=Dict("facecolor"=>"white", "alpha"=>0.9), fontsize=lsize, horizontalalignment="center",verticalalignment="center", transform = ax[6].transAxes);
ax[8].text(0.1, 0.95, L"\mathbf{(h)}", bbox=Dict("facecolor"=>"white", "alpha"=>0.9), fontsize=lsize, horizontalalignment="center",verticalalignment="center", transform = ax[8].transAxes);

# savefig("./drag_figs/LSA_omega_c.pdf",bbox_inches="tight")
```

```julia

function calc_wf(z_ind_vec, h_ind, mu_ind)

    # one for loop to rule them all (psi_ot edition)
    i=h_ind  # bottom slope index
    j=1  # quad drag index
    k=mu_ind  # linear drag index
    
    
    h0 = (h0s[i], 0.0)
    kappa_loc = kappas[j]
    mu_loc = mus[k]
    
    global model_params = redef_mu_kappa_topoPV_h0(model_params, mu_loc, kappa_loc, (0., 0.), h0)
    global topo_PV, eta = define_topo(model_params)
    global model_params = redef_mu_kappa_topoPV_h0(model_params, mu_loc, kappa_loc, topo_PV, h0)
    
    global model_params = redef_mu_kappa_beta(model_params, mu_loc, kappa_loc, 0.)
    
    
    # for z_ind in [40, 100, 190, 285, 370, 455, 550, 630, 710, 785, 880, 970]
    # [70, 145, 235, 315, 410, 505, 590, 670, 742, 840, 925, 1020]

    
    wf = nothing # zeros(Int(size(a["jld_data"]["psi_ot_slice"])[1]/2+1), size(a["jld_data"]["psi_ot_slice"][:,:,:,1:10:end])[4])
    cnt = 0
    
    for z_ind in z_ind_vec
        psi_slice = zeros(1024,0)
        
        for yr in collect(range(45.5, 89.5,step=0.5))
            a = load(data_dir_psi_ot*jld_name(model_params,yr))
    
            layer_no = 1
    
            # psi_BT_0 = 0.5 .* (a["jld_data"]["psi_ot"][:,z_ind,1, :] .+ a["jld_data"]["psi_ot"][:,z_ind,2, :])
            # psi_slice = cat(psi_slice, psi_BT_0, dims=2)
    
            psi_slice = cat(psi_slice, a["jld_data"]["psi_ot"][:,z_ind,layer_no, :], dims=2)
    
            
        end
    
        # psi_slice = psi_slice[:,1:10:end];
    
        if isnothing(wf)
            # psi_hat = rfft(psi_slice .- mean(psi_slice))
            # spec = -grid.kr.^2 .* psi_hat .* conj.(psi_hat)
    
            # wf = abs.(fftshift(spec))
    
            mid_ind = floor(Int, length(psi_slice[1,:])/2)
            wf = 0.5 * (abs.(fftshift(rfft(psi_slice[:, 1:mid_ind] .- mean(psi_slice[:, 1:mid_ind])))) .+ abs.(fftshift(rfft(psi_slice[:, mid_ind+1:Int(2*mid_ind)] .- mean(psi_slice[:, mid_ind:Int(2*mid_ind)])))))
           
        else
            # psi_hat = rfft(psi_slice .- mean(psi_slice))
            # spec = -grid.kr.^2 .* psi_hat .* conj.(psi_hat)
    
            # wf .+= abs.(fftshift(spec))
            mid_ind = floor(Int, length(psi_slice[1,:])/2)
            
            wf .+= 0.5 * (abs.(fftshift(rfft(psi_slice[:, 1:mid_ind] .- mean(psi_slice[:, 1:mid_ind])))) .+ abs.(fftshift(rfft(psi_slice[:, mid_ind+1:Int(2*mid_ind)] .- mean(psi_slice[:, mid_ind:Int(2*mid_ind)])))))
           
        end
    
        cnt+=1
    
        GC.gc()
    
    end
        
    wf = wf ./ cnt;

    return wf
end

    
```

```julia
############################################################################

function define_wavefreqs(mu_loc; kvec_str="zonal")
    ells = kx.^-1
    Nell = length(ells)
    
    c_BT_full = zeros(length(h0s), Nell) .* NaN
    c_BC_full = zeros(length(h0s), Nell) .* NaN
    c_RG = zeros(length(h0s), Nell) .* NaN

    # ks = collect(range(0.001, 2, Nell)) ./ Ld
    ells = kx.^-1

    for h_ind in range(1, length(h0s))
    
        h0 = (h0s[h_ind], 0.0)
        kappa_loc = kappas[1]
        
        global model_params = redef_mu_kappa_topoPV_h0(model_params, mu_loc, kappa_loc, (0., 0.), h0)
        global topo_PV, eta = define_topo(model_params)
        global model_params = redef_mu_kappa_topoPV_h0(model_params, mu_loc, kappa_loc, topo_PV, h0)

        if h_ind != 14
            # a = load(data_dir*jld_name_2L(model_params,62.95))

            # CBCh = a["jld_data"]["two_layer_kspace_modal_nrg_budget"][:,4]

            # if sum(findall(x -> x<0, CBCh[1:20])) >0
            #     neg_max[h_ind] = maximum(findall(x -> x<0, CBCh[1:20]))
            # end

            
            # v_scale_BT =  (sqrt(2*a["jld_data"]["two_layer_xspace_modal_nrg_budget"][1]))

            alpha_s = (f0 * h0s[h_ind] / H[2])

            for (i,L0) in enumerate(ells)
                # if h_ind != 7

                if kvec_str=="zonal"
                    kx=2*pi/L0
                    ky=0
                elseif kvec_str=="ky_fixed"
                    kx = 2*pi/L0
                    ky = 2*pi/Lx*5
                elseif kvec_str=="k_iso"
                    kx = 2*pi/L0
                    ky = 2*pi/L0
                end

                c_BT_full[h_ind, i], nul = dispersion_relation_full_bottom(L0, alpha_s, 0.01, "BT", mu_loc; kx=kx, ky=ky) 
                c_BC_full[h_ind, i], nul = dispersion_relation_full_bottom(L0, alpha_s, 0.01, "BC", mu_loc; kx=kx, ky=ky)

                c_RG[h_ind, i] = dispersion_relation_RG(L0, alpha_s, 0.01; kx=kx, ky=ky) 

                # end
            end
        end
        
    end
        
    return ells, c_BT_full .* (2 * pi ./ ells)', c_BC_full .* (2 * pi ./ ells)', c_RG .* (2 * pi ./ ells)'
    # return ells, c_BT_full ./ ells', c_BC_full ./ ells'

end

es, f_BT_topo_ky0_mu2, f_BC_topo_ky0_mu2, f_RG_ky0_mu2 = define_wavefreqs(mus[2])

es, f_BT_topo_kyjet_mu2, f_BC_topo_kyjet_mu2, f_RG_kyjet_mu2 = define_wavefreqs(mus[2]; kvec_str="ky_fixed")

es, f_BT_topo_iso_mu2, f_BC_topo_iso_mu2, f_RG_iso_mu2 = define_wavefreqs(mus[2]; kvec_str="k_iso")

##
es, f_BT_topo_ky0_mu4, f_BC_topo_ky0_mu4, f_RG_ky0_mu4 = define_wavefreqs(mus[4])

es, f_BT_topo_kyjet_mu4, f_BC_topo_kyjet_mu4, f_RG_kyjet_mu4 = define_wavefreqs(mus[4]; kvec_str="ky_fixed")

es, f_BT_topo_iso_mu4, f_BC_topo_iso_mu4, f_RG_iso_mu4 = define_wavefreqs(mus[4]; kvec_str="k_iso");
```

```julia
function set_thresh(dat, pct_lim)
    masked_dat = deepcopy(dat)
    N = length(dat)
    sum_dat = sum(dat)
    mean_dat = sum_dat / N
    
    thresh_fact = 0.5
    eps = 0.1
    
    while sum(masked_dat) > pct_lim * sum_dat
        thresh_fact += eps 
        mask = dat .< thresh_fact * mean_dat
        masked_dat[mask] .= 0.
    end
    return thresh_fact * mean_dat
end

# plot_var[wf_mask] .= NaN

kx2 = cat(kx[259:end], kx[1:258], dims=1)
# plt.pcolormesh(kx2, -fr, plot_var')

# plt.plot(kx2)

plt.plot(f_BT_topo_ky0[13,:] )

# plt.ylim(-0.005,0.005)
# plt.xlim(kx[1]*Ld * 2 * pi, kx[20]*Ld*2*pi)

# plt.plot(ifftshift(kx))
```

```julia

function define_psi_slice(z_ind, a)
    psi_slice = zeros(1024,0)
    
    for yr in collect(range(63.95, 94.95,step=0.5))
    
        a = load(data_dir_psi_ot*jld_name(model_params,yr))
    
        layer_no = 1
    
        psi_loc = a["jld_data"]["psi_ot"][:,z_ind,layer_no, :]
        # psi_loc = 0.5 .* (a["jld_data"]["psi_ot"][:,z_ind,1, :] .+ a["jld_data"]["psi_ot"][:,z_ind,2, :])
        
        psi_slice = cat(psi_slice, psi_loc, dims=2)
    
    end
    
    
    t_start = size(psi_slice)[2]  # this accounts for early times when we are pushing to t but not saving psi yet
    
    # t = a["jld_data"]["t"][1:10:end][end - t_start + 1: end] ./ (3600 * 24)
    # t = t .- t[1];
    
    t = a["jld_data"]["t"][end - t_start + 1: end] ./ (3600 * 24 * 7)
    t = t .- t[1];
    
    return t, psi_slice
end
```

# Calculating "vertical mode mixing" as function of wavenumber

```julia
A11(c, kx, ky) = (c - 0.01) * ( - (kx^2 + ky^2) - 0.5 * Ld^-2) - 0.5 * 0.01 * Ld^-2
A12(c, kx, ky) = 0.5 * (c - 0.01) * Ld^-2
```
